Importing libraries

```python
import torch
import numpy as np
import pandas as pd
from torchvision.transforms import transforms
from torch.utils.data.sampler import SubsetRandomSampler
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets
```

checking availability of GPU on your machine

```python
gpu_available=torch.cuda.is_available()
```

uploading and transformation of data

```python
nm_of_worker=0
batch_size=20
valid_size=0.2
transformation=transforms.Compose([
                                   transforms.Resize((512,512)),
                                   transforms.RandomHorizontalFlip(),
                                   transforms.ToTensor(),
                                   transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
                                  ])

test_path="/content/drive/My Drive/data/Test"
train_path="/content/drive/My Drive/data/Train"

test_dataset=datasets.ImageFolder(test_path,transform=transformation)
train_dataset=datasets.ImageFolder(train_path,transform=transformation)
```

shuffling of data

```python
nm_test=len(test_dataset)
list_idx_test=list(range(nm_test))
np.random.shuffle(list_idx_test)

nm_train=len(train_dataset)
list_idx_train=list(range(nm_train))
np.random.shuffle(list_idx_train)
```

split the train data and upload on loaders

```python
split=int(np.floor(valid_size*nm_train))

test_idx=list_idx_test
train_idx,valid_idx=list_idx_train[split:],list_idx_train[:split]

train_sampler=SubsetRandomSampler(train_idx)
valid_sampler=SubsetRandomSampler(valid_idx)
test_sampler=SubsetRandomSampler(test_idx)

train_loader=torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,
                                         sampler=train_sampler,num_workers=nm_of_worker)
valid_loader=torch.utils.data.DataLoader(train_dataset,batch_size=batch_size,
                                         sampler=valid_sampler,num_workers=nm_of_worker)
test_loader=torch.utils.data.DataLoader(test_dataset,batch_size=batch_size,
                                         sampler=test_sampler,num_workers=nm_of_worker)
classes = [0,1]
```

function to show an image by unnormalizing it

```python
def imshow(img):
    img = img /2 + 0.5 
    #print(img.size)
   # img=img*255
    plt.imshow(np.transpose(img, (1, 2, 0)))
```

show the images of train loader

```python
dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images.numpy() # convert images to numpy for display
#print(images[0].size)
# plot the images in the batch, along with the corresponding labels
fig = plt.figure(figsize=(25, 4))
# display 20 images
for idx in np.arange(batch_size):
    ax = fig.add_subplot(2, batch_size/2, idx+1, xticks=[], yticks=[])
    imshow(images[idx])
   # print(labels[idx])
    if labels[idx]==0:
      ax.set_title("covid +ve")
    else:
      ax.set_title("covid -ve")
```

Defining the architecture for CNN 

```python
class Net(nn.Module):
   def __init__(self):
     super(Net,self).__init__()
     self.conv1=nn.Conv2d(3,8,3,padding=1)
     self.conv2=nn.Conv2d(8,16,3,padding=1)
     self.pool=nn.MaxPool2d(2,2)
     self.dp1=nn.Dropout(p=0.25,inplace=True)
     self.conv3=nn.Conv2d(16,32,3,padding=1)
     self.conv4=nn.Conv2d(32,64,3,padding=1)
     self.dp2=nn.Dropout(p=0.4,inplace=True)
     self.fc1=nn.Linear(64*64*64,64)
     self.fc2=nn.Linear(64,2) 
     self.softmax = nn.LogSoftmax(dim=1)
   def forward(self,x):
     x=F.relu(self.conv1(x))
     x=F.relu(self.conv2(x))
     x=self.pool(x)
     x=self.dp1(x)
     x=F.relu(self.conv3(x))
     x=self.pool(x)
     x=self.dp1(x)
     x=F.relu(self.conv4(x))
     x=self.pool(x)
     x=self.dp2(x)
     x=x.view(-1,64*64*64)
     x = F.relu(self.fc1(x))
     x = self.dp1(x)
     x = self.softmax(self.fc2(x))
     return x 
model=Net()
if gpu_available:
  model.cuda()
print(model)  
```

selecting the loss function and optimizer function 

```python
criterion  = nn.CrossEntropyLoss(weight=None,size_average=True)
optimizer  = optim.Adam(model.parameters() , lr=0.003)
```

training of data

```python
epochs=50
min_loss=1e18

for e in range(epochs):
  train_loss=0
  valid_loss=0
  model.train()
  for d,t in train_loader:
    if gpu_available:
      d=d.cuda()
      t=t.cuda()
    h=model(d)
    loss=criterion(h,t)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    train_loss+=loss.item()*d.size(0)
  
  model.eval()
  for d,t in valid_loader:
    if gpu_available:
      d=d.cuda()
      t=t.cuda()
    h=model(d)
    loss=criterion(h,t)
    valid_loss+=loss.item()*d.size(0)

  train_loss/= len(train_loader.sampler)
  valid_loss/= len(valid_loader.sampler)
  print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        e, train_loss, valid_loss))
    
  if valid_loss <= min_loss:
    print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
          min_loss,
          valid_loss))
    torch.save(model.state_dict(), 'cur_model.pt')
    min_loss = valid_loss      
```

load the model with min val loss

```python
model.load_state_dict(torch.load('cur_model.pt'))
if gpu_available:
    model = model.cuda()
model
```

test the model for accuracy

```python
test_loss = 0.0
class_correct = list(0.0 for i in range(2) )
class_total = list(0.0 for i in range(2))

model.eval()

for data,target in test_loader:
    if gpu_available:
        data = data.cuda()
        target = target.cuda()
    
    output = model(data)
    loss = criterion(output,target)
    test_loss+= loss.item()*data.size(0)
    
    _ , pred = torch.max(output,dim=1)
    correct_tensor = pred.eq(target.data.view_as(pred))
    correct = np.squeeze(correct_tensor.numpy()) if not gpu_available else np.squeeze(correct_tensor.cpu().numpy())
    for i in range(target.shape[0]):
        label = target.data[i]
        class_correct[label] += correct[i].item()
        class_total[label] += 1

test_loss = test_loss/len(test_loader.dataset)
print('Test Loss: {:.6f}\n'.format(test_loss))

for i in range(2):
    if class_total[i] > 0:
        print('Test Accuracy of %5s: %2d%% (%2d/%2d)' % (
            classes[i], 100 * class_correct[i] / class_total[i],
            np.sum(class_correct[i]), np.sum(class_total[i])))
    else:
        print('Test Accuracy of %5s: N/A (no training examples)' % (classes[i]))

    
print('\nTest Accuracy (Overall): %2d%% (%2d/%2d)' % (
    100. * np.sum(class_correct) / np.sum(class_total),
    np.sum(class_correct), np.sum(class_total)))
```
